```python
# import the library
import pandas as pd
```

```python
raw_data_1 = pd.read_csv('koordinat-dan-ketinggian-kantor-kelurahan-di-kota-bandung-2014.csv')
raw_data_1.head()
```

```python
raw_data_2 = pd.read_csv('tahun-2020-s2---jumlah-kepadatan-penduduk.csv')
raw_data_2.head()
```

```python
data = raw_data_1[['Kecamatan', 'Kelurahan', 'Lintang Selatan', 'Bujur Timur']].rename(columns={'Kecamatan':'District', 'Kelurahan':'Subdistrict', 'Lintang Selatan':'Latitude', 'Bujur Timur':'Longitude'})
data['Population Density'] = 0
for i in range(len(data)):
    for j in range(len(raw_data_2)):
        if data['District'][i] == raw_data_2['Kecamatan'][j]:
            data['Population Density'][i] = raw_data_2['Kepadatan (Jiwa/Ha)'][j]
data
```
